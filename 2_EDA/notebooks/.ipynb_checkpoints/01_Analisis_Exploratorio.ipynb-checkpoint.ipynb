```python
import importlib
import subprocess
import sys

def install_if_missing(package):
    try:
        # Check if the module is available
        importlib.import_module(package.split('-')[0])
    except ImportError:
        print(f"Installing {package}...")
        subprocess.check_call([sys.executable, "-m", "pip", "install", package])

# List of required libraries
required_packages = ["psycopg2-binary", "sqlalchemy"]

for pkg in required_packages:
    install_if_missing(pkg)
```

```python
from sqlalchemy import create_engine
# El host es el nombre del servicio en el docker-compose
engine = create_engine('postgresql://admin:adminpassword@nike_postgres:5432/nike_dw')
```

```python
# Este comando lista todas las tablas del usuario
query = "SELECT table_name FROM information_schema.tables WHERE table_schema = 'raw'"
tablas = pd.read_sql(query, engine)
print("Tablas en la base de datos:")
print(tablas)
```
